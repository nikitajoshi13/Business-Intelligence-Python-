```python
display(Markdown("**a. Load the csv file of dataset into dataframe in Python program.**"))
import pandas as pd
df=pd.read_csv("KCLT.csv")
df.columns
```

```python
display(Markdown("**b. Find mean, median, mode, min and max for all numeric attributes**"))
import numpy as np
display(Markdown("**Mean**"))
for i in df.columns[1:]: #mean is not valid for date,[1:] for columns starting from 1 
    print(i,": ",df[i].mean())
display(Markdown("**Median**"))
for i in df.columns[1:]: #mean is not valid for date,[1:] for columns starting from 1 
    print(i,": ",df[i].median())
display(Markdown("**Mode**"))
for i in df.columns[1:]: #mean is not valid for date,[1:] for columns starting from 1 
    print(i,": ",df[i].mode())
```

```python
display(Markdown("**c. Plot min and max for actual_mean_temp,actual_min_temp and actual_max_temp using bar graph. Use blue for min and red for max.**"))
import numpy as np
import matplotlib.pyplot as plt

n_groups = 3
minimums = (min(df['actual_mean_temp']),min(df['actual_min_temp']),min(df['actual_max_temp']))
maximums = (max(df['actual_mean_temp']),max(df['actual_min_temp']),max(df['actual_max_temp']))
#data to be plotted is ready

#create a plotb=
fig,ax = plt.subplots() #creats just a figure and only 1 subplot
index = np.arange(n_groups)#for labels on X-axis & number of subplots
bar_width = 0.35
opacity = 0.8

#plt.bar(x,height,bar_width)
result1 = plt.bar(index,minimums,bar_width,alpha=opacity,color='b',label='Minimum')
result2 = plt.bar(index+bar_width,maximums,bar_width,alpha=opacity,color='r',label='Maxmum')
#both the subplots have no labels on X-axis or Y-axis
#index_bar_width to distinct subplots in a group

plt.xlabel('Categories of Temperature')
plt.ylabel('Values')
plt.title('Comparison of min, max and mean')
#No labels till now for subplots

plt.xticks(index+bar_width,('Actual_mean_temp','Actual_min_temp','Actual_max_temp'))
# xticks()-Get or set the current tick locations and labels of the x-axis.
#xticks(ticks, [labels], **kwargs)  # Set locations and labels

plt.legend() #for showing what different colors repreesent

#plt.tight_layout()
plt.show()
```

```python
#A histogram represents the frequency distribution of a variable.
import datetime
display(Markdown("**d. Plot two separate histograms for actual_mean_temp for months of Aug 2014 and March 2015. Make key observations (put them as comments in jupyter notebook).**"))

df2 = pd.read_csv("KCLT.csv",parse_dates=['date'])#To deal with dates

cols = [col for col in df2.columns if col in ['date','actual_mean_temp']] #for keeping only particluar columns
df3 = df2[cols]

df4=df3[(df3['date']>pd.Timestamp(2014,7,31)) & (df3['date']<pd.Timestamp(2014,9,1))] #for keeping relevant dates

#plt.hist(x, bins=None, range=None, density=None)# x should be an array
df4.hist(bins=15)
plt.xlabel('Mean Temperature')
plt.ylabel('Probability')
plt.title('Actual Mean Temperature for Aug,2014')
plt.grid(True)

#For March 2015
df4=df3[(df3['date']>pd.Timestamp(2015,2,28)) & (df3['date']<pd.Timestamp(2015,4,1))] #for keeping relevant dates

#plt.hist(x, bins=None, range=None, density=None)# x should be an array
df4.hist(bins=15)
plt.xlabel('Mean Temperature')
plt.ylabel('Probability')
plt.title('Actual Mean Temperature for March,2015')
plt.grid(True)
```

```python
#a simple way of representing statistical data on a plot in which a rectangle is drawn to represent the second and third quartiles,
display(Markdown("**Plot box plots for attributes actual_mean_temp, actual_min_temp and actual_max_temp. Make key observations (put them as comments in jupyter notebook)**"))
cols = [col for col in df.columns if col in ['actual_mean_temp', 'actual_min_temp', 'actual_max_temp']]
df2=df[cols]

boxplot = df2.boxplot(column=['actual_mean_temp', 'actual_min_temp', 'actual_max_temp'])
```

```python
display(Markdown("**f. Find correlation between following pairs of attributes: actual_mean_temp & actual_min_temp and actual_mean_temp & actual_max_temp. Make key observations.**"))

# Between actual_mean_temp & actual_min_temp
print(df["actual_mean_temp"].corr(df["actual_min_temp"]))
display(Markdown("** They are highly correlated**"))
                 
                 
# Between actual_mean_temp & actual_max_temp
print(df["actual_mean_temp"].corr(df["actual_max_temp"]))
display(Markdown("**They are highly correlated**"))
                
display(Markdown("** Both the quantities are almost equally correlated to the mean**"))
```

```python
display(Markdown("**g. Plot two scatter plots between following pairs of attributes: actual_mean_temp & actual_min_temp and actual_mean_temp & actual_max_temp. Make key observations.**"))
df.plot.scatter("actual_mean_temp", "actual_min_temp")
plt.title("Scatter plot b/w actual_mean_temp & actual_min_temp")

df.plot.scatter("actual_mean_temp", "actual_max_temp")
plt.title("Scatter plot b/w actual_mean_temp & actual_max_temp")

display(Markdown("**Both the scatter plots show a linear relationship**"))
```

```python
#Correlation gives an indication of how related the changes are between two variables.We can calculate the correlation between each pair of attributes. This is called a correlation matrix
display(Markdown("**h. Make correlation matrix plot and scatterplot matrix taking all temperature attributes in the dataset.**"))

#cols = [col for col in df.columns if col not in ['date']] #for keeping only particluar columns
#df3 = df[cols]

correlations = df.corr()
fig = plt.figure() #Create a new figure
ax = fig.add_subplot(111) #"111" means "1x1 grid
cax = ax.matshow(correlations, vmin=-1, vmax=1) #Display an array as a matrix in a new figure window.
#min and vmax define the data range that the colormap covers. By default, the colormap covers the complete value range of the supplied data. 
fig.colorbar(cax) #colorbar for different colors
ticks = np.arange(0,12,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
#ax.set_xticklabels(df.columns)
ax.set_yticklabels(df.columns)
plt.show()


#ScatterPLot
from pandas.plotting import scatter_matrix
scatter_matrix(df)
plt.show()
```
